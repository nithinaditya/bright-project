```python
weather_df = pd.read_csv('weather.csv', index_col = [0])
```

```python
weather_df = weather_df[columns]
```

```python
columns = ["windspeedKmph", "winddirDegree", "weatherCode", "precipMM", "visibility", "pressure", "cloudcover", "DewPointF", "WindGustKmph", "tempF", "WindChillF", "humidity", "date", "time", "airport_code"]
```

```python
flights_df['CRSDepTime'] = pd.to_numeric(flights_df['CRSDepTime'], errors='coerce')
flights_df['rounded_time'] = flights_df['CRSDepTime'].apply(lambda time: round(time / 100) * 100)
```

```python
flights_df['rounded_time']
```

```python
final_df = pd.merge(flights_df, weather_df, how = "inner", left_on = ["rounded_time", "FlightDate", "Origin"], right_on=["time", "date", "airport_code"])
```
